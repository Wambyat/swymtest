```python
import sqlite3

db_name = 'test.db'
conn = sqlite3.connect(db_name)

# Assuming df and df_meta are your dataframes
# table_name_df = 'main'
table_name_df_meta = 'main_meta'

# Save df into the database
# columns image,feature,rank,description,also_view,also_buy,similar_item, tech1 (this a dict)
df_meta['image'] = df_meta['image'].fillna({i: [] for i in df_meta.index})
df_meta['image'] = df_meta['image'].apply(json.dumps)
df_meta['feature'] = df_meta['feature'].fillna({i: [] for i in df_meta.index})
df_meta['feature'] = df_meta['feature'].apply(json.dumps)
df_meta['rank'] = df_meta['rank'].fillna({i: {} for i in df_meta.index})
df_meta['rank'] = df_meta['rank'].apply(json.dumps)
df_meta['description'] = df_meta['description'].fillna({i: {} for i in df_meta.index})
df_meta['description'] = df_meta['description'].apply(json.dumps)
df_meta['also_view'] = df_meta['also_view'].fillna({i: [] for i in df_meta.index})
df_meta['also_view'] = df_meta['also_view'].apply(json.dumps)
df_meta['also_buy'] = df_meta['also_buy'].fillna({i: [] for i in df_meta.index})
df_meta['also_buy'] = df_meta['also_buy'].apply(json.dumps)
df_meta['similar_item'] = df_meta['similar_item'].fillna({i: [] for i in df_meta.index})
df_meta['similar_item'] = df_meta['similar_item'].apply(json.dumps)
df_meta['tech1'] = df_meta['tech1'].fillna({i: {} for i in df_meta.index})
df_meta['tech1'] = df_meta['tech1'].apply(json.dumps)



# df['style'] = df['style'].fillna({i: {} for i in df.index})
# df['style'] = df['style'].apply(json.dumps)
# df['image'] = df['image'].fillna({i: [] for i in df.index})
# df['image'] = df['image'].apply(json.dumps)
df_meta.to_sql(table_name_df_meta, conn, if_exists='replace', index=False)

# Save df_meta into the database
# df_meta.to_sql(table_name_df_meta, conn, if_exists='replace', index=False)

# Close the database connection
conn.close()
```
