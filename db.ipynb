```python
import sqlite3

# Create or connect to an SQLite database
db_name = 'test.db'
conn = sqlite3.connect(db_name)

# Assuming df and df_meta are your dataframes
table_name_df = 'main'
# table_name_df_meta = 'main_meta'

# Save df into the database
df['style'] = df['style'].fillna({i: {} for i in df.index})
df['style'] = df['style'].apply(json.dumps)
df['image'] = df['image'].fillna({i: [] for i in df.index})
df['image'] = df['image'].apply(json.dumps)
df.to_sql(table_name_df, conn, if_exists='replace', index=False)

# Save df_meta into the database
# df_meta.to_sql(table_name_df_meta, conn, if_exists='replace', index=False)

# Close the database connection
conn.close()
```
